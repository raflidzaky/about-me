## Filtering Tabel

```python
# Multiple Filtering
  ## AND
df4 = df3.loc[(df3['normalized-losses'] != '?') & (df3['num-of-doors'] == 'two')]
df4
  ## OR
df4 = df3.loc[(df3['normalized-losses'] != '?') | (df3['num-of-doors'] == 'two')]
df4
```

## Filtering Tabel: Regex

```python
# Filter contain suatu komponen
  ## in case lupa the full name
df4 = df3.loc[(df3['body-style'].str.contains('edan')) | (df3['body-style'].str.contains('verti'))]
df4
  ### Logikanya, kita mau SELECT * FROM df3 dmn body style mengandung kata edan atau verti

# Code di atas sangatlah ga clean. Solusinya, kita import fungsi regexp untuk cleaning the code with exact function
  ## Regex powerful asf even in SQL too
import re
df4 = df3.loc[df3['body-style'].str.contains('edan|verti', regex = True)]
df4

# Case kalau kita ga ingin dua value dgn contains, gimana hayo?
import re 
df4 = df3.loc[~df3['body-style'].str.contains('edan|verti', regex = True)]
df4
  # Logikanya, kita ingin SELECT dgn kondisi NOT contains edan or verti

# in case kita lupa capitalization gmn dan gamau regexp match nya case sensitive
import re
df4 = df3.loc[df3['body-style'].str.contains('eDan|verTi', flags = re.I, regex = True)]
df4
  ## Logikanya, regex matching itu case sensitive, thus eDan dan verTi udh pasti ga bakal match
  ## makanya, kita mau ga case sensitive matchingnya
```

```python
## Filtering Regex II
  ### Mencari awalan kata
import re
df4 = df3.loc[df3['body-style'].str.contains('^sE', flags = re.I, regex = True)]
df4
    ### Maksud code: Menampilkan df3 dgn bodystyle ber-value awalan Se (case insensitive)
  
  ### Mencari akhiran kata
import re
df4 = df3.loc[df3['body-style'].str.contains('aN*', flags = re.I, regex = True)]
df4 

  ### Mencari awalan kata dgn akhiran bebas
import re
df4 = df3.loc[df3['body-style'].str.contains('sE[dan]*', flags = re.I, regex = True)]
df4 
    #### Maksud code: Menampilkan df3 dgn bodystyle ber-value awalan Se, berakhiran dan (case insensitive)
```
