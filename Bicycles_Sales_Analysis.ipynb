## Data Cleaning

```python
# Drop the rows containing null value
df2 = df1.dropna(axis = 0)
df2.isnull().sum()
df2.info()
```

```python
# Convert standard cost column value into float
  ## Remove the dollar sign
df2['standard_cost'] = df2['standard_cost'].str.replace("$", "")
df2['standard_cost'] = df2['standard_cost'].str.replace(".", "")
  ## Convert the decimal sign from ',' to '.'
df2['standard_cost'] = df2['standard_cost'].str.replace(",", ".").astype('float')

# Convert list price column value into float
  ## Replace ',' to '.'
df2['list_price'] = df2['list_price'].str.replace(',', '.')
df2['list_price'] = df2['list_price'].astype('float')
```

# Exploratory Data Analysis

## Product Sold

```python
# Overall Product Sold
df3 = df2['brand'].loc[df2['order_status'] != 'Cancelled']

for brand in df3.unique():
    product_sold = df3.loc[df2['brand'] == brand].count()
    print(brand, product_sold)
    percentage = product_sold / df3.count()
    print(brand, percentage)
    print('--------')

print("=====================")
# Overall Product Cancelled
df4 = df2['brand'].loc[df2['order_status'] == 'Cancelled']

for brand in df4.unique():
    product_cancelled = df4.loc[df2['brand'] == brand].count()
    print(brand, product_cancelled)
    percentage = product_cancelled / df4.count()
    print(brand, percentage)
    print('--------')
```

## Transaction Behavior

```python
# Compare online or offline behavior
online = df2['online_order'].loc[df2['online_order'] == True]
offline = df2['online_order'].loc[df2['online_order'] == False]

print(f'those who do online shopping: {online.count()}')
print(f'those who do offline shopping: {offline.count()}')
```

```python
# Lets find out what product class of each brand that got cancelled
cancelled_brands = df2[['brand','product_class']].loc[df2['order_status'] == 'Cancelled']
cancelled_brands_counts = cancelled_brands.value_counts(ascending=True)
print(cancelled_brands_counts)

print("=============")
approved_brands = df2[['brand','product_class']].loc[df2['order_status'] == 'Approved']
approved_brands_counts = approved_brands.value_counts(ascending=True)
print(approved_brands_counts)
```

```python
# Let's see what transactions kind got cancelled
cancelled = df2['online_order'].loc[df2['order_status'] == 'Cancelled']
transactions_mode_cancelled = cancelled.value_counts()
print(transactions_mode_cancelled)

print('-------')
# How each brand's transaction?
transactions_cancelled = df2[['online_order','brand']].loc[df2['order_status'] == 'Cancelled']
transactions_cancelled = transactions_cancelled.value_counts(ascending = True)
print(transactions_cancelled)
```

## Pricing

```python
# What product that has highest price
## By Average
brand = ['Solex', 'Giant Bicycles', 'Trek Bicycles', 'OHM Cycles', 'WeareA2B', 'Norco Bicycles']
product_class = ['low', 'medium', 'high']

for i in brand:
    print(i)
    brand_data = df2[df2['brand'] == i]  # Filter data by its brand
    average_prices = brand_data[['product_class','list_price']].groupby('product_class').mean()  # Count the mean
    print(average_prices)
    print('==============')
```

```python
# For comparison, I also display central tendency of prices by its median
brand = ['Solex', 'Giant Bicycles', 'Trek Bicycles', 'OHM Cycles', 'WeareA2B', 'Norco Bicycles']
product_class = ['low', 'medium', 'high']

for i in brand:
    print(i)
    brand_data = df2[df2['brand'] == i]
    average_prices = brand_data[['product_class','list_price']].groupby('product_class').median()
    print(average_prices)
    print('==============')
```
